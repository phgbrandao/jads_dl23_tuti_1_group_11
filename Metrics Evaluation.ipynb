```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#PRECESSING UNIQUE TOKEN

with open('training_log_uni.txt','r',encoding='utf-16-le') as open_file:
    model_uni = open_file.readlines()
    model_uni = [line.strip().split() for line in model_uni]


epoch = [ int(line[7:][1]) for line in model_uni if len(line)>7 and line[7]=='epoch']
epoch_loss = [ float(line[7:][3]) for line in model_uni if len(line)>7 and line[7]=='epoch']
eval_loss = [ float(line[7:][2]) for line in model_uni if len(line)>7 and line[7]=='eval_loss'][:-1]
eval_acc = [ float(line[7:][2]) for line in model_uni if len(line)>7 and line[7]=='eval_acc'][:-1]
eval_TP = [ float(line[7:][2]) for line in model_uni if len(line)>7 and line[7]=='eval_TP'][:-1]
eval_FP = [ float(line[7:][2]) for line in model_uni if len(line)>7 and line[7]=='eval_FP'][:-1]
eval_TN = [ float(line[7:][2]) for line in model_uni if len(line)>7 and line[7]=='eval_TN'][:-1]
eval_FN = [ float(line[7:][2]) for line in model_uni if len(line)>7 and line[7]=='eval_FN'][:-1]

print(len(epoch), len(epoch_loss),len(eval_loss), len(eval_acc),len(eval_TP))

train_table_uni = pd.DataFrame( {'epoch':epoch,'epoch_loss':epoch_loss,'eval_loss':eval_loss,'eval_acc':eval_acc,
                             'eval_TP':eval_TP,'eval_FP':eval_FP,'eval_TN':eval_TN,'eval_FN':eval_FN})

train_table_uni['recall'] = train_table_uni.eval_TP / (train_table_uni.eval_TP+train_table_uni.eval_FN)
train_table_uni['precision'] = train_table_uni.eval_TP / (train_table_uni.eval_TP+train_table_uni.eval_FP)
train_table_uni['f1_score'] = 2*train_table_uni.recall*train_table_uni.precision / (train_table_uni.recall+train_table_uni.precision)

train_table_uni.head()
```

```python
#PRECESSING INDEX TOKEN

with open('training_log_idx.txt','r',encoding='utf-8') as open_file:
    model_idx = open_file.readlines()
    model_idx = [line.strip().split() for line in model_idx]

poch = [ int(line[7:][1]) for line in model_idx if len(line)>7 and line[7]=='epoch']
epoch_loss = [ float(line[7:][3]) for line in model_idx if len(line)>7 and line[7]=='epoch']
eval_loss = [ float(line[7:][2]) for line in model_idx if len(line)>7 and line[7]=='eval_loss'][:-1]
eval_acc = [ float(line[7:][2]) for line in model_idx if len(line)>7 and line[7]=='eval_acc'][:-1]
eval_TP = [ float(line[7:][2]) for line in model_idx if len(line)>7 and line[7]=='eval_TP'][:-1]
eval_FP = [ float(line[7:][2]) for line in model_idx if len(line)>7 and line[7]=='eval_FP'][:-1]
eval_TN = [ float(line[7:][2]) for line in model_idx if len(line)>7 and line[7]=='eval_TN'][:-1]
eval_FN = [ float(line[7:][2]) for line in model_idx if len(line)>7 and line[7]=='eval_FN'][:-1]

print(len(epoch), len(epoch_loss),len(eval_loss), len(eval_acc),len(eval_TP))

train_table_idx = pd.DataFrame( {'epoch':epoch,'epoch_loss':epoch_loss,'eval_loss':eval_loss,'eval_acc':eval_acc,
                             'eval_TP':eval_TP,'eval_FP':eval_FP,'eval_TN':eval_TN,'eval_FN':eval_FN})

train_table_idx['recall'] = train_table_idx.eval_TP / (train_table_idx.eval_TP+train_table_idx.eval_FN)
train_table_idx['precision'] = train_table_idx.eval_TP / (train_table_idx.eval_TP+train_table_idx.eval_FP)
train_table_idx['f1_score'] = 2*train_table_idx.recall*train_table_idx.precision / (train_table_idx.recall+train_table_idx.precision)

train_table_idx.head()
```

### ROC - Threshold impact

```python
with open('predictions_idx.txt','r') as open_file:
    pred_idx = open_file.readlines()
    pred_idx = [ [float (item) for item in line.strip().split()] for line in pred_idx ]


header = ['idx','true_value','predict_percentage','predict']
pred_idx_dict = [ dict(zip(header,line)) for line in pred_idx]
pred_idx_dict[:5]

thr_idx = []
TPR_idx = []
FPR_idx = []

for threshold in np.arange(0, 1, .05):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for item in pred_idx_dict:
        if item['true_value'] == 1:
            if item['predict_percentage'] >= threshold:
                TP += 1
            else:
                FN += 1
        else:
            if item['predict_percentage'] >= threshold:
                FP += 1
            else:
                TN += 1

    TPR_idx.append( TP/(TP+FN) )
    FPR_idx.append( FP / (FN+FP))
    thr_idx.append(threshold)
```

```python
with open('predictions_uni.txt','r') as open_file:
    pred_uni = open_file.readlines()
    pred_uni = [ [float (item) for item in line.strip().split()] for line in pred_uni ]


header = ['uni','true_value','predict_percentage','predict']
pred_uni_dict = [ dict(zip(header,line)) for line in pred_uni]
pred_uni_dict[:5]

thr_uni = []
TPR_uni = []
FPR_uni = []

for threshold in np.arange(0, 1, .05):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for item in pred_uni_dict:
        if item['true_value'] == 1:
            if item['predict_percentage'] >= threshold:
                TP += 1
            else:
                FN += 1
        else:
            if item['predict_percentage'] >= threshold:
                FP += 1
            else:
                TN += 1

    TPR_uni.append( TP/(TP+FN) )
    FPR_uni.append( FP / (FN+FP))
    thr_uni.append(threshold)
```

```python
plt.figure(figsize=(10,10))
plt.plot([0, 1], [0, 1],ls="--")
plt.plot(FPR_uni,TPR_uni, label = 'Model Unique')


for i, txt in enumerate(thr_uni):
    plt.annotate(round(txt,2), (FPR_uni[i], TPR_uni[i]))

plt.plot(FPR_idx,TPR_idx, label = 'Model Index')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE TEST DATA - Unique & Index')

for i, txt in enumerate(thr_idx):
    plt.annotate(round(txt,2), (FPR_idx[i], TPR_idx[i]))

plt.legend()
plt.show()
```
